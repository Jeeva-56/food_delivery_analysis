# Import Modules

```python
import pandas as pd
from sqlalchemy import create_engine, text
```

# Read CSV

```python
pd.set_option("display.max_columns", 30)
```

```python
df = pd.read_csv("../csv_files/online_food_delivery_feature.csv")
df.head()
```

# Connecting to Mysql using SQLAlchemy

```python
# creating configuration parameters 
username = "root"
password = "root"
host = "localhost"
db_name = "food_delivery_analysis"
```

```python
# connect to mysql using sqlalchemy
engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}"
)
```

```python
# Creating a new database 
conn = engine.connect()

try:
    conn.execute(text(
        f"CREATE DATABASE IF NOT EXISTS {db_name};"
    ))
    print(f"\"{db_name}\" Database Created Successfully.")
except:
    print("Error Occured While Creating Database.")
```

```python
# Reconnecting with database

engine = create_engine(
    f"mysql+pymysql://{username}:{password}@{host}/{db_name}"
)

conn = engine.connect()
```

# Create a Table

```python
query = """
CREATE TABLE IF NOT EXISTS food_delivery
(
    Order_ID                 VARCHAR(50) PRIMARY KEY,
    Customer_ID              VARCHAR(50),
    Customer_Age             INT,
    Customer_Gender          VARCHAR(15),
    City                     VARCHAR(30),
    Area                     VARCHAR(30),
    Restaurant_ID            VARCHAR(20),
    Restaurant_Name          VARCHAR(30),
    Cuisine_Type             VARCHAR(20),
    Order_Date               DATETIME,
    Delivery_Time_Min        INT,
    Distance_km              DECIMAL(5, 2),
    Order_Value              DECIMAL(7, 2),
    Discount_Applied         DECIMAL(7, 2),
    Final_Amount             DECIMAL(7, 2),
    Payment_Mode             VARCHAR(20),
    Order_Status             VARCHAR(20),
    Cancellation_Reason      VARCHAR(20),
    Delivery_Partner_ID      VARCHAR(20),
    Delivery_Rating          INT,
    Restaurant_Rating        DECIMAL(2, 1),
    Order_Day                VARCHAR(20),
    Peak_Hour                INT,
    Profit_Margin            DECIMAL(4, 2),
    Delivery_performance_cat VARCHAR(20),
    Cus_Age_Group            VARCHAR(20),
    Ord_Day                  INT,
    Ord_Month                INT,
    Ord_Year                 INT
);
"""

try:
    conn.execute(text(query))
    conn.commit()

    print("Table Created Successfull!")
except:
    print("Error Occured while creating Table.")
    
```

### Inserting Records into MySQL

```python
df.to_sql(
    name="food_delivery",
    con=engine,
    if_exists="append",
    index=False
)

conn.commit()
```

### Terminating MySQL Connection

```python
conn.close()
```
